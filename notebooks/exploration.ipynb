```python
import importlib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../dataset/dataset_cleaned.csv")
print(df['source'].value_counts())
df.head()
```

```python
print(df.dtypes)
print("\nNull values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8,4))
df['type'].value_counts().plot(kind='bar', color='steelblue')
plt.title('Distribution of Text Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['tone'].value_counts().plot(kind='bar', color='coral')
plt.title('Distribution of Tones')
plt.xlabel('Tone')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
df['source'].value_counts().plot(kind='bar', color='mediumseagreen')
plt.title('Distribution of Sources')
plt.xlabel('Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
